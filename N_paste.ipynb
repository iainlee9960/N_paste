```python
def num_len(num):
    num_str = str(num)

    num_digits = len(num_str)

    return num_digits
```

```python
import matplotlib.pyplot as plt

def plot(x, y):
    # Plotting the graph
    y = list(map(lambda x: num_len(x), y))
    plt.plot(x, y, marker='o')

    # Adding labels and title to the graph
    plt.xlabel('X Values')
    plt.ylabel('Y Values')
    plt.title('Graph of X and Y Values')

    # Displaying the graph
    plt.show()
```

```python
plot(x, y)
plot(x2, y2)
plot(x3, y3)
plot(x4, y4)
plot(x5, y5)
plot(x6, y6)
plot(x7, y7)
plot(x8, y8)
plot(x9, y9)
```
